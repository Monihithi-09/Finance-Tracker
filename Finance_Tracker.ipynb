```python
import pandas as pd

df = pd.read_excel("Copy of personal_transactions_dashboard_ready (2).xlsx")

print(df.head())
```

```python
print(df.info())
```

# Cleaning the dataset

```python
df['Date'] = pd.to_datetime(df['Date'])
df=df.dropna()
df=df.drop_duplicates()
df = df[df['Amount'] > 0]
print(df)
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
print(df)
```

```python
print(df.info())
```

## Standardize catogory names

```python
df['Description'] = df['Description'].astype(str)
df['Description'] = df['Description'].str.lower().str.strip()

def categorize(expense):
    expense = str(expense)

    if any(word in expense for word in ['food', 'restaurant', 'coffee', 'grocery', 'starbucks']):
        return 'food'

    elif any(word in expense for word in ['mortgage', 'bill', 'electric', 'water', 'internet', 'rent', 'insurance', 'credit card']):
        return 'bills'

    elif any(word in expense for word in ['transport', 'uber', 'bus', 'taxi', 'train', 'gas', 'bp', 'shell']):
        return 'travel'

    elif any(word in expense for word in ['amazon', 'flipkart', 'shopping', 'clothing']):
        return 'shopping'

    else:
        return 'entertainment'

df['main_category'] = df['Description'].apply(categorize)

print(df[['Description', 'main_category']].head(20))
print(df['main_category'].value_counts())
```
